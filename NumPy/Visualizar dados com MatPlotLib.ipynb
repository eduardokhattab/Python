```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
salario = np.array([100,200,300,500,400,150])
salario
```

```python
plt.plot(salario)
```

```python
plt.plot(salario, c="red") # c de color
```

```python
plt.plot(salario, c="salmon", ls='--') #line style
```

```python
plt.plot(salario, ls='dashed', marker='o') # o de bolinha
```

```python
salarios_marcos = np.array([100,200,400,300])
salarios_gileno = np.array([50,300,500,450])
```

```python
plt.plot(salarios_marcos, c='red', ls='dashed', marker='o')
plt.plot(salarios_gileno, c='blue', ls='dashed', marker='o')
```

```python
# adicionar label + legenda
plt.plot(salarios_marcos, c='red', ls='dashed', marker='o', label='Salários Marcos')
plt.plot(salarios_gileno, c='blue', ls='dashed', marker='o', label='Salários Gileno')

plt.legend(loc='lower right') # legenda com as labels, loc é a posicao
plt.show() # remove <matplotlib.lines.Line2D at 0xa851ed0>
```
